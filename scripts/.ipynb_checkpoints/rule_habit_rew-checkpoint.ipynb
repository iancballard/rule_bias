# Load all data files

```python
sub_df = []
order = {}
for sub in subs:
    files = glob.glob(op.join(data_dir, sub, '*'))
    files = [x for x in files if "reward" in x]
    
    for n,f in enumerate(files):
        block_type = f.split('/')[-1].split('_')[-2]
        block_num = int(f.split('/')[-1].split('_')[-1].split('.')[0])
                
        with open(f, 'rb') as f:
            p = pickle.load(f)
            ntrials = len(p.rt)
            
            df = pd.DataFrame({'rule': p.active_rule,
                   'rt':p.rt,
                   'resp':p.resp,
                   'correct_resp':p.correct_resp,
                   'correct':p.correct,
                    'sub':sub,
                    'cond':block_type,
                    'trial': np.arange(ntrials),
                    'block': block_type,
                    'miniblock': p.miniblock,
                    'block_num': block_num,
                    'rew':p.rew,
                    'rewarded_rule': p.rewarded_rule,
                  'color_coherence': p.coherences['color'][:ntrials],
                  'motion_coherence': p.coherences['motion'][:ntrials],
                  'shape_coherence': p.coherences['shape'][:ntrials]})
            sub_df.append(df.copy())
    print(sub,p.coherence_floor)
    
df = pd.concat(sub_df)
df = df.sort_values(by = ['sub','block','block_num','trial'])
df = df.set_index('sub')
```

```python
#set some derived variables
df['logrt'] = [np.log(x) for x in df['rt']]

df['test_type'] = df['rule'] == df['rewarded_rule']
test_map = {True: 'rewarded', False: 'nonrewarded'}
df['test_type'] = [test_map[x] for x in df['test_type']]
```

# set up miniblocks

```python
#get miniblock type
for sub in subs:
    miniblock_type = []
    for n,row in df.loc[sub].iterrows():
        tt = row['rewarded_rule']
        miniblock = row['miniblock']
        rule = row['rule']
        if tt in miniblock:
            if rule == tt:
                miniblock_type.append('rewarded')
            else:
                miniblock_type.append('compete')
        else:
            miniblock_type.append('noncompete')
    df.loc[sub,'miniblock_type'] = miniblock_type
```

```python
#index miniblock trial
miniblock_trial = []
mb = df['miniblock'].values
counter = 0
for n,val in enumerate(mb):
    if n>0:
        if val == mb[n-1]:
            counter +=1
        else:
            counter = 0
    if counter > 11:
        counter = 0
    miniblock_trial.append(counter)
df['miniblock_trial'] = miniblock_trial
```

```python
#switch/stay
for sub in subs:

    switch = ['switch']
    rules = df.loc[sub,'rule'].values
    for n,r in enumerate(rules):
        if n > 0:
            if r ==rules[n-1]:
                switch.append('stay')
            else:
                switch.append('switch')
    df.loc[sub,'switch'] = switch
```

```python
#switch type and error handling
for sub in subs:
    sub_df = df.loc[sub].copy().reset_index()

    switch_type = ['switch_other']    
    error_last_trial = ['correct']

    for n,row in sub_df.iterrows():
        
        #switch/stay
        rule = row['rule']
        switch = row['switch']
        tt = row['rewarded_rule']

        if n > 0:        
            if switch == 'switch':
                if rule == tt:
                    switch_type.append('switch_to')
                elif sub_df.loc[n - 1,'rule'] == tt:
                    switch_type.append('switch_away')
                else:
                    switch_type.append('switch_other')
            else:
                switch_type.append('stay')
        
            #error last trial
            if sub_df.loc[n - 1,'correct']:
                error_last_trial.append('correct')
            else:
                error_last_trial.append('incorrect')
    
    df.loc[sub,'prev_trial_acc'] = error_last_trial       
    df.loc[sub,'switch_type'] = switch_type
```

```python
#bin the coherences
for sub in subs:
    for rule in ['color','motion','shape']:

        a = list(set(df.loc[sub][rule + '_coherence']))
        a.sort()
        coherence_map = {x:n for n,x in enumerate(a)}
        df.loc[sub,rule + '_coherence_bin'] = [coherence_map[x] for x in df.loc[sub][rule + '_coherence']]
```

```python
#get coherence level for each trial
for sub in subs:
    coherence = []
    coherence_bin = []
    test_type = []
    for n,row in df.loc[sub].iterrows():

        #get coherence for chosen rule
        if row['rule'] == 'motion':
            coherence.append(row['motion_coherence'])
            coherence_bin.append(row['motion_coherence_bin'])

        if row['rule'] == 'color':
            coherence.append(row['color_coherence'])
            coherence_bin.append(row['color_coherence_bin'])

        if row['rule'] == 'shape':
            coherence.append(row['shape_coherence'])
            coherence_bin.append(row['shape_coherence_bin'])
        
    df.loc[sub,'coherence'] = coherence
    df.loc[sub,'coherence_bin'] = coherence_bin
```

## Balanced difference analysis

```python
diff_df = []
d = df.reset_index().set_index(['sub','block','block_num']).sort_index()
for sub, block, n in set(d.index):
    sub_df = d.loc[(sub,block,n)]
    sub_df = sub_df.reset_index().set_index(['rule','miniblock_type'])

    #determine rules for this subject
    rules= ['color','motion','shape']
    rewarded_rule = sub_df['rewarded_rule'].values[0]

    for rule in rules:
        if rule == rewarded_rule:
#             diff = sub_df.loc[rule].groupby('switch').mean()
            diff = sub_df.loc[rule].mean()
            cond = 'rewarded'
            
        else:
#             a = sub_df.loc[rule,'compete'].groupby('switch').mean()
#             b = sub_df.loc[rule,'noncompete'].groupby('switch').mean()
            a = sub_df.loc[rule,'compete'].mean()
            b = sub_df.loc[rule,'noncompete'].mean()

            cond = 'compete_minus_noncompete'
            diff = a-b


            
        diff['block'] = block
        diff['block_num'] = n
        diff['sub'] = sub
        diff['rule'] = rule
        diff['cond'] = cond
        
#         diff_df.append(diff.reset_index().copy())
        diff_df.append(pd.DataFrame(diff.to_dict(), index = [0]))

diff_df = pd.concat(diff_df)
```

```python
fig_df = diff_df[diff_df.cond == 'compete_minus_noncompete']
sns.catplot(col = 'sub',
           y="correct",
            row = 'block',
              join = False,
              dodge = .25,
            kind = 'point',
           data=diff_df)
```

# Analyze training data

```python
sub_df = []
for sub in subs:
    files = glob.glob(op.join(data_dir, sub, '*'))
    files = [x for x in files if "training" in x]
    files = [x for x in files if "retrain" not in x]
    
    for n,f in enumerate(files):

        block_type = f.split('/')[-1].split('_')[-3].split('.')[0]
        block_num = int(f.split('/')[-1].split('_')[-2].split('.')[0])
                
        with open(f, 'rb') as f:
            p = pickle.load(f)
            ntrials = len(p.rt)
            
            df = pd.DataFrame({'rule': p.training_step,
                   'rt':p.rt,
                   'resp':p.resp,
                   'correct_resp':p.correct_resp,
                    'sub':sub,
                    'cond':block_type,
                    'trial': np.arange(ntrials),
                    'block': block_type,
                    'block_num': block_num,
                  'coherence': p.coherence_record[p.training_step][:ntrials]})
            sub_df.append(df.copy())
    
day1 = pd.concat(sub_df)
day1 = day1.sort_values(by = ['sub','block_num','trial'])
day1['correct'] = day1['correct_resp'] == day1['resp']
day1['block_num'] = [math.floor(x/3) for x in day1['block_num']]
day1['grand_trial'] = [x*40 + y for x,y in zip(day1['block_num'],day1['trial'])]
# df = df.set_index(['sub','cond','block_num'])
```
